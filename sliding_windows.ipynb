Maximum Sum Subarray of Size K
nums = [2, 1, 5, 1, 3, 2]
k = 3

```
class sliding_windows:
  def max_sum_subarray(self, nums, k):
    n = len(nums)
    total_sum = 0

    for i in range(k):
      total_sum = total_sum + nums[i]

    maxx =  total_sum

    for i in range(k,n):
      total_sum = total_sum + nums[i] - nums[i - k]

      if total_sum > maxx:
        maxx = total_sum

    return maxx
nums = [2,1,5,1,3,2]
k  = 3

leetcode = sliding_windows()
print(leetcode.max_sum_subarray(nums, k))


```

First Negative Number in Every Window
nums = [12, -1, -7, 8, -15, 30, 16, 28]
k = 3

output = [-1, -1, -7, -15, -15, 0]

```
class sliding_windows:
  def first_negative(self, nums, k):
    n = len(nums)
    result = []
    negatives = []

    for i in range(n):
      if nums[i] < 0:
        negatives.append(i)
      if negatives and negatives[0] <= i - k:
        negatives.pop(0)
      if i >= k - 1:
        if negatives:
          result.append(nums[negatives[0]])
        else:
          result.append(0)


    return result



nums = [12, -1, -7, 8, -15, 30, 16, 28]
k = 3

leetcode = sliding_windows()
print(leetcode.first_negative(nums, k))
```
